```python
##libraries needed
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Fixing columns


```python
soccer_df = pd.read_csv('SalaryPrediction.csv')
soccer_df['Wage'] = soccer_df['Wage'].str.replace(',','').astype(int)
soccer_df['Wage'] = soccer_df['Wage']/1000
soccer_df

sns.set(font_scale=2)
```

## Boxplot of each league

```python
plt.figure(figsize=(20, 10))
sorted_salaries = soccer_df.sort_values(by= 'Wage' ,ascending= False)
top_3_salaries_removed = sorted_salaries.iloc[3:]
sns.boxplot(x='League',y='Wage',data=top_50_salaries_removed).set(title="Box and whisker plot of Salary by League",ylabel = "Wages in thousands of dollars")

```

## average salary by age

```python
plt.figure(figsize=(20, 10))
age_df = soccer_df.groupby('Age').mean()
sns.lineplot(data=age_df)
plt.plot()

```

## Salary By Position


```python
plt.figure(figsize=(20, 10))
top_50_salaries_removed = sorted_salaries.iloc[3:]
sns.boxplot(x='Position',y='Wage',data=top_50_salaries_removed)
```

## Top Average Salary By Club

```python
plt.figure(figsize=(20, 10))
Club_df = soccer_df.groupby('Club').mean()
Club_df = Club_df.sort_values(by =['Wage'],ascending=False)
Club_df_top_20 = Club_df.head(20)
sns.scatterplot(data = Club_df, x="Age", y="Wage").set(title='Clubs Average Salary vs Average Age')
```
