```python
##libraries needed
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Fixing columns


```python
soccer_df = pd.read_csv('SalaryPrediction.csv')
soccer_df['Wage'] = soccer_df['Wage'].str.replace(',','').astype(int)
soccer_df['Wage'] = soccer_df['Wage']/1000



soccer_df
```

## Boxplot of each league

```python
premiera_league = soccer_df['League'] == 'Primiera Liga'
big_5 = soccer_df[~premiera_league]

plt.figure(figsize=(15, 15))
sorted_salaries = big_5.sort_values(by= 'Wage' ,ascending= False)
top_3_salaries_removed = sorted_salaries.iloc[3:]
ax = sns.violinplot(x='League',y='Wage',data=top_3_salaries_removed)
ax.set(ylabel = "Wages in Thousands of Dollars")
ax.yaxis.label.set(fontweight='bold')
ax.xaxis.label.set(fontweight='bold')


plt.savefig('BoxplotLeague.png')

```

## average salary by age

```python
plt.figure(figsize=(15, 15))
age_df = soccer_df.groupby('Age').mean()
bx = sns.lineplot(data=age_df['Wage'])
bx.set(ylabel = "Wages in Thousands of Dollars")
bx.yaxis.label.set(fontweight='bold')
bx.xaxis.label.set(fontweight='bold')
plt.plot()

plt.savefig('Linegraphbyage.png')

```

## Salary By Position


```python
plt.figure(figsize=(15, 15))
top_3_salaries_removed = sorted_salaries.iloc[3:]
cx = sns.violinplot(x='Position',y='Wage',data=top_3_salaries_removed)
cx.set(ylabel = "Wages in Thousands of Dollars")
cx.yaxis.label.set(fontweight='bold')
cx.xaxis.label.set(fontweight='bold')
plt.savefig('BoxplotPosition.png')
```

## Top Average Salary By Club

```python
plt.figure(figsize=(15, 15))
Club_df = soccer_df.groupby('Club').mean()
Club_df = Club_df.sort_values(by =['Wage'],ascending=False)
Club_df_top_20 = Club_df.head(20)
dx = sns.scatterplot(data = Club_df, x="Age", y="Wage")
dx.set(ylabel = "Wages in Thousands of Dollars")
dx.yaxis.label.set(fontweight='bold')
dx.xaxis.label.set(fontweight='bold')
plt.savefig('clubaverage.png')
```
